```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(start=-10, stop=10, num=400)  
'''
np.linspace(start, stop, num) generates num evenly spaced samples, calculated
over the interval [start, stop].
'''
```

```python
x
```

```python
m = 2 
c = 3  
y = m * x + c
```

```python
plt.plot(x, y, label='y = 2x + 3')
plt.title('Plot of the Linear Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() #new one
plt.savefig('matplotlib_linear_equation.png')

'''plt.savefig() allows you to save the figure in various file formats such as PNG, JPEG, SVG, PDF, and more.'''
```

For a quadratic equation

```python
# Define the coefficients
a = 1
b = -4
c = 4

# Generate x values
x = np.linspace(-5, 5, 400)

# Calculate y values using the quadratic equation
y = a * x**2 + b * x + c

# Create the plot
plt.plot(x, y, label='y = x^2 - 4x + 4')

# Adding title and labels
plt.title('Plot of the Quadratic Equation')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-10,10)
plt.ylim(0,50)

plt.legend()
# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

The **zorder** parameter determines the drawing order of plot elements.

```python
# Plot with zorder
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], color='red', linewidth=5, zorder=1)  # Drawn first
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='blue', linewidth=5, zorder=2)  # Drawn on top
plt.plot([1, 2, 3, 4], [10, 13, 15, 17], color='green', linewidth=5, zorder=3) # Drawn on top of blue

# Show plot
plt.show()
```

Bar Plot

```python
x=['A','B','C','D','E']
y=[5,7,3,8,6]

##create a bar plot
plt.bar(x,y,color='purple')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x=['A','B','C','D','E']
y=[5,7,3,8,6]

##create a bar plot
plt.barh(x,y,color='purple')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 7]
values2 = [5, 3, 2]

# Creating stacked bar graph
plt.bar(categories, values1, label='Part 1')
plt.bar(categories, values2, bottom=values1, label='Part 2')

# Adding title, labels, and legend
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

# Show plot
plt.show()
```

Histogram

Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data=[2,5,3,6,7,7,8,1,3,9]

plt.hist(data,bins=10,color='green',edgecolor='white')
```

Pie Chart

```python
labels=['A','B','C','D']
sizes=[30,20,15,35]
colors=['red','green','blue','yellow']
explode=(0.1,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,explode=explode)
```

```python
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```
